### #6

```python
#group by user and calculate the mean price of products purchased
user_avg_prices = ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()
```

```python
#create the spending flag based on the average price
user_avg_prices['spending_flag'] = user_avg_prices['prices'].apply(lambda x: 'Low spender' if x < 10 else 'High spender')
```

```python
#test results
user_avg_prices.head(20)
```

```python
#merge the spending flag back into the original DataFrame
ords_prods_merge = pd.merge(ords_prods_merge, user_avg_prices[['user_id', 'spending_flag']], on='user_id', how='left')
```

```python
#test results
ords_prods_merge.head(20)
```

### #7

```python
#group by user and calculate the median days since prior order
user_order_frequency = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median().reset_index()
```

```python
#create the order frequency flag based on the median days since prior order
def order_frequency_flag(median_days):
    if median_days > 20:
        return 'Non-frequent customer'
    elif median_days > 10:
        return 'Regular customer'
    else:
        return 'Frequent customer'

user_order_frequency['order_frequency_flag'] = user_order_frequency['days_since_prior_order'].apply(order_frequency_flag)
```

```python
#test results
user_order_frequency.head()
```

```python
#merge the order frequency flag back into the original DataFrame
ords_prods_merge = pd.merge(ords_prods_merge, user_order_frequency[['user_id', 'order_frequency_flag']], on='user_id', how='left')
```

```python
#test results
ords_prods_merge[['user_id', 'order_frequency_flag']].head(20)
```

```python
#test results
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
